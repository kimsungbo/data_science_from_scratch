## matplotlib

```python
from matplotlib import pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# x축 연도, y축 gdp 선그래프
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')

# 제목
plt.title("Nominal GDP")

# y축 레이블 추가
plt.ylabel("Billions of $")
plt.show()
```

## 막대 그래프

#### 이산적인 항목들ㄹ에 대한 변화를 보여줄때 사용용이

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "andhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

xs = [i for i, _ in enumerate(movies)]

plt.bar(xs, num_oscars)
plt.ylabel("# of Academy Awards")
plt.title("My Favorite Movies")

plt.xticks([i for i, _ in enumerate(movies)], movies)
plt.show()
```

```python
from collections import Counter

# histogram = 정해진 구간에 해당되는 항목의 개수를 보여줌으로써 값의 분포를 관찰할수 있는 그래프 형태

grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
decile = lambda grade: grade //10 * 10
histogram = Counter(decile(grade) for grade in grades)

plt.bar(histogram.keys(), histogram.values(), 8)

# x축은 -5 ~ 105, y축은 0 ~ 5
plt.axis([-5, 105, 0, 5])

# x축 레이블은 0 부터 100까지 10단위로
plt.xticks([10 * i for i in range(11)])
plt.xlabel("Decile")
plt.ylabel("# of Students")
plt.title("Distribution of Exam 1 Grades")
plt.show()
```

```python
# y축이 0에서 시작하지 않으면 오해를 불러일으키기 쉬움

mentions = [500, 505]
years = [2013, 2014]

plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("# of times I heard someone say 'data science'")

plt.ticklabel_format(useOffset=False)

plt.axis([2012.5, 2014.5, 499, 506])
plt.title("Look at the 'Huge' Increase!")
plt.show()
```

```python
mentions = [500, 505]
years = [2013, 2014]

plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ylabel("# of times I heard someone say 'data science'")

plt.ticklabel_format(useOffset=False)

plt.axis([2012.5, 2014.5, 499, 506])
plt.title("Look at the 'Huge' Increase!")

# 더 적합한 축을 사용하여 더 합리적인 그래프
plt.axis([2012.5, 2014.5, 0, 550])
plt.title("Not So Huge Anymore")
plt.show()
```

## 선그래프

```python
variance = [1, 2, 4, 8, 16, 32,64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]

total_error = [x + y for x, y in zip(variance, bias_squared)]

xs = [i for i, _ in enumerate(variance)]

plt.plot(xs, variance, 'g-', label='variance') # 초록색 실선
plt.plot(xs, bias_squared, 'r-', label='bias^2') # 붉은색 점선
plt.plot(xs, total_error, 'b:', label='total_error') # 파란색 점선

plt.legend(loc=9) # 중앙위쪽 의미
plt.xlabel("model complexity")
plt.title("The Bias-Variance Tradeoff")
plt.show()
```

## 산점도

#### 두 변수간의 연관 관계를 보여주고 싶을때 적합

```python
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

for label, friend_count, minute_count in zip(labels, friends, minutes): 
    plt.annotate(label, xy=(friend_count, minute_count), xytext=(5, -5), textcoords='offset points')
    
plt.title("Daily inutes vs. Number of Friends")
plt.xlabel("# of friends")
plt.ylabel("daily minutes spend on the site")
plt.show()
```

```python
# 자동으로 축의 범위를 설정하게 되면 공장한 비교 불가

test_1_grades = [99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]

plt.scatter(test_1_grades, test_2_grades)
plt.title("Axes Aren't Comparable")
plt.xlabel("test 1 grade")
plt.ylabel("test 2 grade")
plt.show()
```

```python
test_1_grades = [99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]

plt.scatter(test_1_grades, test_2_grades)
plt.title("Axes Aren't Comparable")
plt.xlabel("test 1 grade")
plt.ylabel("test 2 grade")

# plt.axis("equal") 을 사용하면 공정한 비교 가능
plt.axis("equal")
plt.show()
```
